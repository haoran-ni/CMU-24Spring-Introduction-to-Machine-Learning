```python
import matplotlib.pyplot as plt
import decision_tree as dt
import numpy as np
```

```python
head, X_train, Y_train = dt.readData('handout/heart_train.tsv')
head, X_test, Y_test = dt.readData('handout/heart_test.tsv')

M = X_train.shape[1]

train_err_list = []
test_err_list = []

for m in range(M):
    
    # Train decision tree
    tree = dt.DecisionTree(m, head)
    tree.train(X_train, Y_train)
    
    # Predict labels
    pred_train = tree.predict(X_train)
    pred_test = tree.predict(X_test)

    # Store metrics
    train_error = dt.errorRate(Y_train, pred_train)
    test_error = dt.errorRate(Y_test, pred_test)
    train_err_list.append(train_error)
    test_err_list.append(test_error)

# Plot
fig, ax = plt.subplots()
ax.plot(train_err_list, '.-', label="Training error", color='b')
ax.plot(test_err_list, '.-', label="Test error", color='r')
ax.set_xlabel("Depth of decision tree")
ax.set_ylabel("Error rate")
ax.legend()
plt.show()
    
```
